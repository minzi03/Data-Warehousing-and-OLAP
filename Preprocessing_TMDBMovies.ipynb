- Trước 1980: Số lượng phim ít hơn và có thể không đầy đủ hoặc có nhiều năm không có dữ liệu đáng kể, nên có thể ít ý nghĩa hơn cho phân tích hiện đại.
Nếu mục tiêu của bạn là tập trung vào các xu hướng hiện đại, tập trung vào dữ liệu từ khoảng năm 1980 trở đi sẽ là hợp lý nhất.

```python
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')
```

```python
# Lọc dữ liệu từ năm 1980 trở đi
data = data[data['release_date'].dt.year >= 1980]
```

```python
data.isnull().sum()
```

```python
# Các cột chuỗi sẽ được thay thế bằng 'None' hoặc chuỗi rỗng
data['title'] = data['title'].fillna('Unknown')
data['status'] = data['status'].fillna('None')
data['release_date'] = data['release_date'].fillna('Unknown')
data['adult'] = data['adult'].fillna(False)
data['homepage'] = data['homepage'].fillna('None')
data['imdb_id'] = data['imdb_id'].fillna('None')
data['original_language'] = data['original_language'].fillna('None')
data['tagline'] = data['tagline'].fillna('None')
data['genres'] = data['genres'].fillna('None')
data['production_companies'] = data['production_companies'].fillna('None')
data['production_countries'] = data['production_countries'].fillna('None')
data['spoken_languages'] = data['spoken_languages'].fillna('None')

# Các cột số thay thế NaN bằng 0 hoặc giá trị hợp lý
data['vote_average'] = data['vote_average'].fillna(0)
data['vote_count'] = data['vote_count'].fillna(0)
data['revenue'] = data['revenue'].fillna(0)
data['runtime'] = data['runtime'].fillna(0)
data['budget'] = data['budget'].fillna(0)
data['popularity'] = data['popularity'].fillna(0)
```

```python
# Nếu cột 'budget' có giá trị 0, thay thế bằng giá trị trung vị của budget
median_budget = data.loc[data['budget'] > 0, 'budget'].median()
data['budget'] = data['budget'].replace(0, median_budget)

# Lọc doanh thu (revenue) hợp lý
# data = data[((data['revenue'] >= 0) & (data['status'] != 'Released')) | ((data['revenue'] != 0) & (data['status'] == 'Released'))]

# Tính lợi nhuận (profit) từ revenue và budget
# data['profit'] = data['revenue'] - data['budget']
```

```python
# Chuyển genres, production_companies, production_countries, và spoken_languages thành danh sách
data['genres'] = data['genres'].apply(lambda x: [genre.strip() for genre in x.split(',') if genre.strip() != 'None'])
data['production_companies'] = data['production_companies'].apply(lambda x: [comp.strip() for comp in x.split(',') if comp.strip() != 'None'])
data['production_countries'] = data['production_countries'].apply(lambda x: [country.strip() for country in x.split(',') if country.strip() != 'None'])
data['spoken_languages'] = data['spoken_languages'].apply(lambda x: [lang.strip() for lang in x.split(',') if lang.strip() != ''])
```

```python
# Thay thế các giá trị 0 trong 'runtime' bằng giá trị trung vị của runtime
median_runtime = data.loc[data['runtime'] > 0, 'runtime'].median()
data['runtime'] = data['runtime'].replace(0, median_runtime)

# Định nghĩa các khoảng cho 'runtime'
bins = [0, 60, 120, 180, float('inf')]
labels = ['Under 1h', '1h - 2h', '2h - 3h', 'Over 3h']

# Áp dụng pd.cut() để chia cột 'runtime' thành các khoảng
data['runtime_range'] = pd.cut(data['runtime'], bins=bins, labels=labels, right=False)
```

```python
# Phân loại độ phổ biến (popularity)
popularity_data = data['popularity']
popularity_bins = [0, 0.6, 0.874, 10, 100, float('inf')]
popularity_labels = ['Low', 'Average', 'Good', 'Very Good', 'Excellent']

# Phân chia dữ liệu theo các khoảng
data['popularity_category'] = pd.cut(popularity_data, bins=popularity_bins, labels=popularity_labels, include_lowest=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# =============================================================================
# DATA QUALITY CHECK AFTER PREPROCESSING
# =============================================================================

print("="*60)
print("DATA QUALITY ASSESSMENT AFTER PREPROCESSING")
print("="*60)

# 1. BASIC DATA INFORMATION
print("\n1. BASIC DATA INFORMATION")
print("-"*40)
print(f"Dataset shape: {data.shape}")
print(f"Memory usage: {data.memory_usage().sum() / 1024**2:.2f} MB")
print(f"Date range: {data['release_date'].min()} to {data['release_date'].max()}")

# 2. MISSING VALUES CHECK
print("\n2. MISSING VALUES CHECK")
print("-"*40)
missing_values = data.isnull().sum()
missing_percentage = (missing_values / len(data)) * 100

missing_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing_Count': missing_values.values,
    'Missing_Percentage': missing_percentage.values
})
missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

if missing_df.empty:
    print("✓ No missing values found!")
else:
    print("⚠️ Missing values detected:")
    print(missing_df.to_string(index=False))

# 3. DATA TYPES VALIDATION
print("\n3. DATA TYPES VALIDATION")
print("-"*40)
expected_types = {
    'id': ['int64', 'float64'],
    'title': ['object'],
    'vote_average': ['float64', 'int64'],
    'vote_count': ['int64', 'float64'],
    'status': ['object'],
    'release_date': ['datetime64[ns]'],
    'revenue': ['int64', 'float64'],
    'runtime': ['float64', 'int64'],
    'adult': ['bool', 'object'],
    'budget': ['float64', 'int64'],
    'homepage': ['object'],
    'imdb_id': ['object'],
    'original_language': ['object'],
    'popularity': ['float64', 'int64'],
    'tagline': ['object'],
    'genres': ['object'],
    'production_companies': ['object'],
    'production_countries': ['object'],
    'spoken_languages': ['object']
}

print("Data types validation:")
for col, expected in expected_types.items():
    if col in data.columns:
        actual_type = str(data[col].dtype)
        status = "✓" if actual_type in expected else "⚠️"
        print(f"{status} {col}: {actual_type}")

# 4. NUMERICAL COLUMNS VALIDATION
print("\n4. NUMERICAL COLUMNS VALIDATION")
print("-"*40)

numerical_cols = ['vote_average', 'vote_count', 'revenue', 'runtime', 'budget', 'popularity']

for col in numerical_cols:
    if col in data.columns:
        print(f"\n{col.upper()}:")
        print(f"  Range: {data[col].min():.2f} - {data[col].max():.2f}")
        print(f"  Mean: {data[col].mean():.2f}")
        print(f"  Median: {data[col].median():.2f}")
        print(f"  Std: {data[col].std():.2f}")
        
        # Check for negative values (except revenue which can be 0)
        negative_count = (data[col] < 0).sum()
        if negative_count > 0:
            print(f"  ⚠️ Negative values: {negative_count}")
        else:
            print(f"  ✓ No negative values")
        
        # Check for zero values
        zero_count = (data[col] == 0).sum()
        print(f"  Zero values: {zero_count} ({zero_count/len(data)*100:.1f}%)")

# 5. CATEGORICAL COLUMNS VALIDATION
print("\n5. CATEGORICAL COLUMNS VALIDATION")
print("-"*40)

categorical_cols = ['status', 'original_language', 'adult', 'runtime_range', 'popularity_category']

for col in categorical_cols:
    if col in data.columns:
        print(f"\n{col.upper()}:")
        value_counts = data[col].value_counts()
        print(f"  Unique values: {data[col].nunique()}")
        print(f"  Most common: {value_counts.index[0]} ({value_counts.iloc[0]} records)")
        
        # Show top 5 values
        print("  Top 5 values:")
        for i, (value, count) in enumerate(value_counts.head().items()):
            print(f"    {i+1}. {value}: {count} ({count/len(data)*100:.1f}%)")

# 6. DATE VALIDATION
print("\n6. DATE VALIDATION")
print("-"*40)

if 'release_date' in data.columns:
    print(f"Date range: {data['release_date'].min()} to {data['release_date'].max()}")
    
    # Check for future dates
    future_dates = data[data['release_date'] > pd.Timestamp.now()]
    print(f"Future release dates: {len(future_dates)} movies")
    
    # Check date distribution by year
    data['release_year'] = data['release_date'].dt.year
    year_dist = data['release_year'].value_counts().sort_index()
    print(f"Year range: {year_dist.index.min()} - {year_dist.index.max()}")
    print(f"Most productive year: {year_dist.idxmax()} ({year_dist.max()} movies)")

# 7. LIST COLUMNS VALIDATION
print("\n7. LIST COLUMNS VALIDATION")
print("-"*40)

list_cols = ['genres', 'production_companies', 'production_countries', 'spoken_languages']

for col in list_cols:
    if col in data.columns:
        print(f"\n{col.upper()}:")
        
        # Check if all values are lists
        try:
            list_lengths = data[col].apply(len)
            print(f"  Average items per movie: {list_lengths.mean():.2f}")
            print(f"  Max items: {list_lengths.max()}")
            print(f"  Movies with no items: {(list_lengths == 0).sum()}")
            
            # Get all unique items
            all_items = []
            for items in data[col]:
                all_items.extend(items)
            unique_items = len(set(all_items))
            print(f"  Total unique items: {unique_items}")
            
        except Exception as e:
            print(f"  ⚠️ Error processing list column: {e}")

# 8. BUSINESS LOGIC VALIDATION
print("\n8. BUSINESS LOGIC VALIDATION")
print("-"*40)

# Revenue vs Budget consistency
if 'revenue' in data.columns and 'budget' in data.columns:
    high_revenue_low_budget = data[(data['revenue'] > 100000000) & (data['budget'] < 1000000)]
    print(f"High revenue (>100M) with low budget (<1M): {len(high_revenue_low_budget)} movies")
    
    # Movies with budget > revenue (potential losses)
    losses = data[data['budget'] > data['revenue']]
    print(f"Movies with budget > revenue: {len(losses)} ({len(losses)/len(data)*100:.1f}%)")

# Vote average validation
if 'vote_average' in data.columns:
    invalid_votes = data[(data['vote_average'] < 0) | (data['vote_average'] > 10)]
    print(f"Invalid vote_average (not 0-10): {len(invalid_votes)} movies")

# Runtime validation
if 'runtime' in data.columns:
    very_short = data[data['runtime'] < 60]
    very_long = data[data['runtime'] > 300]
    print(f"Very short movies (<60 min): {len(very_short)}")
    print(f"Very long movies (>300 min): {len(very_long)}")

# 9. OUTLIER DETECTION
print("\n9. OUTLIER DETECTION")
print("-"*40)

for col in numerical_cols:
    if col in data.columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
        print(f"{col}: {len(outliers)} outliers ({len(outliers)/len(data)*100:.1f}%)")

# 10. DATA COMPLETENESS SCORE
print("\n10. DATA COMPLETENESS SCORE")
print("-"*40)

# Calculate completeness score
total_cells = data.shape[0] * data.shape[1]
non_null_cells = total_cells - data.isnull().sum().sum()
completeness_score = (non_null_cells / total_cells) * 100

print(f"Overall data completeness: {completeness_score:.2f}%")

# Essential columns completeness
essential_cols = ['id', 'title', 'release_date', 'budget', 'revenue']
essential_completeness = []

for col in essential_cols:
    if col in data.columns:
        completeness = ((data[col].notna()).sum() / len(data)) * 100
        essential_completeness.append(completeness)
        print(f"{col} completeness: {completeness:.2f}%")

if essential_completeness:
    avg_essential_completeness = np.mean(essential_completeness)
    print(f"Essential columns average completeness: {avg_essential_completeness:.2f}%")

# 11. VISUALIZATION OF DATA QUALITY
print("\n11. GENERATING DATA QUALITY VISUALIZATIONS")
print("-"*40)

# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Missing values heatmap
if not missing_df.empty:
    sns.heatmap(data.isnull(), cbar=True, ax=axes[0,0])
    axes[0,0].set_title('Missing Values Heatmap')
else:
    axes[0,0].text(0.5, 0.5, 'No Missing Values', ha='center', va='center', transform=axes[0,0].transAxes)
    axes[0,0].set_title('Missing Values Status')

# Distribution of key numerical columns
data[['budget', 'revenue', 'runtime', 'vote_average']].hist(bins=20, ax=axes[0,1])
axes[0,1].set_title('Distribution of Key Numerical Columns')

# Release date distribution
if 'release_year' in data.columns:
    data['release_year'].value_counts().sort_index().plot(kind='line', ax=axes[1,0])
    axes[1,0].set_title('Movies Count by Release Year')
    axes[1,0].set_xlabel('Year')
    axes[1,0].set_ylabel('Number of Movies')

# Budget vs Revenue scatter plot
if 'budget' in data.columns and 'revenue' in data.columns:
    sample_data = data.sample(min(1000, len(data)))  # Sample for performance
    axes[1,1].scatter(sample_data['budget'], sample_data['revenue'], alpha=0.6)
    axes[1,1].set_xlabel('Budget')
    axes[1,1].set_ylabel('Revenue')
    axes[1,1].set_title('Budget vs Revenue (Sample)')
    axes[1,1].set_xscale('log')
    axes[1,1].set_yscale('log')

plt.tight_layout()
plt.show()

# 12. FINAL SUMMARY
print("\n" + "="*60)
print("DATA QUALITY SUMMARY")
print("="*60)

quality_issues = []

if not missing_df.empty:
    quality_issues.append(f"Missing values in {len(missing_df)} columns")

if any(data[col].min() < 0 for col in numerical_cols if col in data.columns):
    quality_issues.append("Negative values in numerical columns")

if completeness_score < 95:
    quality_issues.append(f"Low completeness score ({completeness_score:.1f}%)")

if quality_issues:
    print("⚠️ QUALITY ISSUES DETECTED:")
    for issue in quality_issues:
        print(f"  - {issue}")
else:
    print("✓ DATA QUALITY LOOKS GOOD!")

print(f"\nDataset is ready for analysis with {data.shape[0]} records and {data.shape[1]} features.")
print("="*60)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# =============================================================================
# DATA QUALITY COMPARISON: ORIGINAL vs PREPROCESSED
# =============================================================================

print("="*80)
print("DATA QUALITY IMPROVEMENT ANALYSIS")
print("="*80)

# Load original data for comparison
print("Loading original dataset...")
original_data = pd.read_csv('TMDB_movie_dataset_v11.csv')

# Current preprocessed data is already in 'data' variable
preprocessed_data = data.copy()

print(f"Original dataset shape: {original_data.shape}")
print(f"Preprocessed dataset shape: {preprocessed_data.shape}")

# =============================================================================
# 1. MISSING VALUES COMPARISON
# =============================================================================

print("\n" + "="*60)
print("1. MISSING VALUES IMPROVEMENT")
print("="*60)

# Get common columns only
common_columns = list(set(original_data.columns) & set(preprocessed_data.columns))
print(f"Common columns: {len(common_columns)}")

# Calculate missing values for common columns only
original_missing_common = original_data[common_columns].isnull().sum()
original_missing_pct_common = (original_missing_common / len(original_data)) * 100

preprocessed_missing_common = preprocessed_data[common_columns].isnull().sum()
preprocessed_missing_pct_common = (preprocessed_missing_common / len(preprocessed_data)) * 100

# Create comparison dataframe for common columns
missing_comparison = pd.DataFrame({
    'Column': common_columns,
    'Original_Missing': original_missing_common.values,
    'Original_Missing_%': original_missing_pct_common.values,
    'Preprocessed_Missing': preprocessed_missing_common.values,
    'Preprocessed_Missing_%': preprocessed_missing_pct_common.values
})

# Calculate improvement
missing_comparison['Improvement_%'] = missing_comparison['Original_Missing_%'] - missing_comparison['Preprocessed_Missing_%']

print("Missing Values Comparison (Common Columns):")
print("-" * 80)
for _, row in missing_comparison.iterrows():
    if row['Original_Missing_%'] > 0 or row['Preprocessed_Missing_%'] > 0:
        print(f"{row['Column']:<25} | Original: {row['Original_Missing_%']:>6.1f}% | "
              f"Preprocessed: {row['Preprocessed_Missing_%']:>6.1f}% | "
              f"Improvement: {row['Improvement_%']:>6.1f}%")

# Overall missing values improvement for common columns
original_total_missing = original_missing_common.sum()
preprocessed_total_missing = preprocessed_missing_common.sum()

original_missing_rate = (original_total_missing / (len(original_data) * len(common_columns))) * 100
preprocessed_missing_rate = (preprocessed_total_missing / (len(preprocessed_data) * len(common_columns))) * 100

missing_improvement = original_missing_rate - preprocessed_missing_rate

print(f"\nOVERALL MISSING VALUES IMPROVEMENT (Common Columns):")
print(f"Original missing rate: {original_missing_rate:.2f}%")
print(f"Preprocessed missing rate: {preprocessed_missing_rate:.2f}%")
print(f"Improvement: {missing_improvement:.2f} percentage points")

# =============================================================================
# 2. DATA COMPLETENESS COMPARISON
# =============================================================================

print("\n" + "="*60)
print("2. DATA COMPLETENESS IMPROVEMENT")
print("="*60)

# Calculate completeness scores for common columns
original_completeness = ((len(original_data) * len(common_columns) - original_total_missing) / 
                        (len(original_data) * len(common_columns))) * 100

preprocessed_completeness = ((len(preprocessed_data) * len(common_columns) - preprocessed_total_missing) / 
                            (len(preprocessed_data) * len(common_columns))) * 100

completeness_improvement = preprocessed_completeness - original_completeness

print(f"Original data completeness: {original_completeness:.2f}%")
print(f"Preprocessed data completeness: {preprocessed_completeness:.2f}%")
print(f"Completeness improvement: {completeness_improvement:.2f} percentage points")

# =============================================================================
# 3. DATA QUALITY ISSUES COMPARISON
# =============================================================================

print("\n" + "="*60)
print("3. DATA QUALITY ISSUES COMPARISON")
print("="*60)

# Check numerical columns for quality issues (only common ones)
numerical_cols = ['vote_average', 'vote_count', 'revenue', 'runtime', 'budget', 'popularity']
numerical_cols_common = [col for col in numerical_cols if col in common_columns]

quality_issues_original = 0
quality_issues_preprocessed = 0

print("Numerical Data Quality Issues:")
print("-" * 50)

for col in numerical_cols_common:
    # Convert to numeric, errors='coerce' will turn non-numeric to NaN
    orig_col = pd.to_numeric(original_data[col], errors='coerce')
    prep_col = pd.to_numeric(preprocessed_data[col], errors='coerce')
    
    # Count issues in original data
    orig_negative = (orig_col < 0).sum()
    orig_zero = (orig_col == 0).sum()
    orig_null = orig_col.isnull().sum()
    
    # Count issues in preprocessed data  
    prep_negative = (prep_col < 0).sum()
    prep_zero = (prep_col == 0).sum()
    prep_null = prep_col.isnull().sum()
    
    orig_issues = orig_negative + orig_null
    prep_issues = prep_negative + prep_null
    
    quality_issues_original += orig_issues
    quality_issues_preprocessed += prep_issues
    
    print(f"{col}:")
    print(f"  Original   - Negative: {orig_negative:>5}, Null: {orig_null:>5}, Zero: {orig_zero:>5}")
    print(f"  Preprocessed - Negative: {prep_negative:>5}, Null: {prep_null:>5}, Zero: {prep_zero:>5}")
    
    if orig_issues > prep_issues:
        improvement = orig_issues - prep_issues
        print(f"  ✓ Improved by {improvement} issues")
    elif orig_issues < prep_issues:
        regression = prep_issues - orig_issues
        print(f"  ⚠️ Added {regression} issues")
    else:
        print(f"  → No change")

# =============================================================================
# 4. DATE QUALITY IMPROVEMENT
# =============================================================================

print("\n" + "="*60)
print("4. DATE QUALITY IMPROVEMENT")
print("="*60)

if 'release_date' in common_columns:
    # Original date issues
    orig_date_null = original_data['release_date'].isnull().sum()
    
    # Try to convert original dates
    orig_dates_converted = pd.to_datetime(original_data['release_date'], errors='coerce')
    orig_date_invalid = orig_dates_converted.isnull().sum() - orig_date_null
    
    # Preprocessed date issues
    prep_date_null = preprocessed_data['release_date'].isnull().sum()
    prep_date_invalid = 0  # Assuming preprocessing handled invalid dates
    
    print(f"Date Quality Issues:")
    print(f"Original data - Null dates: {orig_date_null}, Invalid dates: {orig_date_invalid}")
    print(f"Preprocessed data - Null dates: {prep_date_null}, Invalid dates: {prep_date_invalid}")
    
    date_improvement = (orig_date_null + orig_date_invalid) - (prep_date_null + prep_date_invalid)
    print(f"Date quality improvement: {date_improvement} issues resolved")

# =============================================================================
# 5. FEATURE ENGINEERING ADDITIONS
# =============================================================================

print("\n" + "="*60)
print("5. FEATURE ENGINEERING ADDITIONS")
print("="*60)

original_features = len(original_data.columns)
preprocessed_features = len(preprocessed_data.columns)
new_features = preprocessed_features - len(common_columns)

print(f"Original features: {original_features}")
print(f"Preprocessed features: {preprocessed_features}")
print(f"Common features: {len(common_columns)}")
print(f"New features added: {new_features}")

# List new features
new_feature_names = set(preprocessed_data.columns) - set(common_columns)
if new_feature_names:
    print(f"New features created: {', '.join(sorted(new_feature_names))}")

# =============================================================================
# 6. OVERALL DATA QUALITY SCORE
# =============================================================================

print("\n" + "="*60)
print("6. OVERALL DATA QUALITY SCORE")
print("="*60)

# Calculate quality scores (0-100)
# Factors: completeness, valid values, consistency

# Original quality score
if len(numerical_cols_common) > 0:
    orig_quality_score = (
        original_completeness * 0.4 +  # 40% weight for completeness
        (100 - (quality_issues_original / (len(original_data) * len(numerical_cols_common)) * 100)) * 0.4 +  # 40% for valid values
        50 * 0.2  # 20% base score for consistency (assumed moderate)
    )

    # Preprocessed quality score
    prep_quality_score = (
        preprocessed_completeness * 0.4 +  # 40% weight for completeness
        (100 - (quality_issues_preprocessed / (len(preprocessed_data) * len(numerical_cols_common)) * 100)) * 0.4 +  # 40% for valid values
        90 * 0.2  # 20% for consistency (assumed high after preprocessing)
    )
else:
    orig_quality_score = original_completeness * 0.8 + 50 * 0.2
    prep_quality_score = preprocessed_completeness * 0.8 + 90 * 0.2

quality_improvement_score = prep_quality_score - orig_quality_score

print(f"Original Data Quality Score: {orig_quality_score:.1f}/100")
print(f"Preprocessed Data Quality Score: {prep_quality_score:.1f}/100")
print(f"Quality Improvement: {quality_improvement_score:.1f} points")

# =============================================================================
# 7. VISUALIZATION OF IMPROVEMENTS
# =============================================================================

print("\n" + "="*60)
print("7. GENERATING IMPROVEMENT VISUALIZATIONS")
print("="*60)

# Create comparison visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Missing values comparison
missing_cols = missing_comparison[missing_comparison['Original_Missing_%'] > 0]['Column'].head(10)
if len(missing_cols) > 0:
    x_pos = np.arange(len(missing_cols))
    orig_vals = missing_comparison[missing_comparison['Column'].isin(missing_cols)]['Original_Missing_%']
    prep_vals = missing_comparison[missing_comparison['Column'].isin(missing_cols)]['Preprocessed_Missing_%']
    
    axes[0,0].bar(x_pos - 0.2, orig_vals, 0.4, label='Original', color='red', alpha=0.7)
    axes[0,0].bar(x_pos + 0.2, prep_vals, 0.4, label='Preprocessed', color='green', alpha=0.7)
    axes[0,0].set_xlabel('Columns')
    axes[0,0].set_ylabel('Missing Values %')
    axes[0,0].set_title('Missing Values Improvement')
    axes[0,0].set_xticks(x_pos)
    axes[0,0].set_xticklabels(missing_cols, rotation=45, ha='right')
    axes[0,0].legend()
else:
    axes[0,0].text(0.5, 0.5, 'No Missing Values to Compare', ha='center', va='center', transform=axes[0,0].transAxes)
    axes[0,0].set_title('Missing Values Status')

# 2. Completeness comparison
categories = ['Original', 'Preprocessed']
completeness_scores = [original_completeness, preprocessed_completeness]
colors = ['red', 'green']

axes[0,1].bar(categories, completeness_scores, color=colors, alpha=0.7)
axes[0,1].set_ylabel('Completeness %')
axes[0,1].set_title('Data Completeness Improvement')
axes[0,1].set_ylim(0, 100)

# Add value labels on bars
for i, v in enumerate(completeness_scores):
    axes[0,1].text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom')

# 3. Quality score comparison
quality_scores = [orig_quality_score, prep_quality_score]
axes[1,0].bar(categories, quality_scores, color=colors, alpha=0.7)
axes[1,0].set_ylabel('Quality Score')
axes[1,0].set_title('Overall Quality Score Improvement')
axes[1,0].set_ylim(0, 100)

# Add value labels on bars
for i, v in enumerate(quality_scores):
    axes[1,0].text(i, v + 1, f'{v:.1f}', ha='center', va='bottom')

# 4. Feature count comparison
feature_counts = [original_features, preprocessed_features]
axes[1,1].bar(categories, feature_counts, color=['blue', 'orange'], alpha=0.7)
axes[1,1].set_ylabel('Number of Features')
axes[1,1].set_title('Feature Engineering Results')

# Add value labels on bars
for i, v in enumerate(feature_counts):
    axes[1,1].text(i, v + 0.5, f'{v}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# =============================================================================
# 8. FINAL IMPROVEMENT SUMMARY
# =============================================================================

print("\n" + "="*80)
print("FINAL DATA QUALITY IMPROVEMENT SUMMARY")
print("="*80)

print(f"📊 DATASET SIZE:")
print(f"   • Records: {len(original_data):,} → {len(preprocessed_data):,} ({((len(preprocessed_data) - len(original_data))/len(original_data)*100):+.1f}%)")
print(f"   • Features: {original_features} → {preprocessed_features} (+{new_features})")

print(f"\n📈 QUALITY METRICS:")
print(f"   • Data Completeness: {original_completeness:.1f}% → {preprocessed_completeness:.1f}% (+{completeness_improvement:.1f}%)")
print(f"   • Missing Values: {original_missing_rate:.1f}% → {preprocessed_missing_rate:.1f}% (-{missing_improvement:.1f}%)")
print(f"   • Overall Quality Score: {orig_quality_score:.1f} → {prep_quality_score:.1f} (+{quality_improvement_score:.1f} points)")

print(f"\n🔧 KEY IMPROVEMENTS:")
improvements = []
if missing_improvement > 0:
    improvements.append(f"✓ Reduced missing values by {missing_improvement:.1f} percentage points")
if quality_improvement_score > 0:
    improvements.append(f"✓ Improved quality score by {quality_improvement_score:.1f} points")
if new_features > 0:
    improvements.append(f"✓ Added {new_features} new engineered features")
if completeness_improvement > 0:
    improvements.append(f"✓ Increased data completeness by {completeness_improvement:.1f}%")

if improvements:
    for improvement in improvements:
        print(f"   {improvement}")
else:
    print("   • No significant improvements detected")

print(f"\n📋 PREPROCESSING EFFECTIVENESS:")
if quality_improvement_score >= 20:
    effectiveness = "EXCELLENT"
elif quality_improvement_score >= 10:
    effectiveness = "GOOD"
elif quality_improvement_score >= 5:
    effectiveness = "MODERATE"
else:
    effectiveness = "LIMITED"

print(f"   • Overall Effectiveness: {effectiveness}")
print(f"   • Quality Improvement: {quality_improvement_score:.1f}% increase")

print("\n" + "="*80)
```
