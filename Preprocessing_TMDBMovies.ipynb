```python
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')
```

```python
# Lá»c dá»¯ liá»‡u tá»« nÄƒm 1980 trá»Ÿ Ä‘i
data = data[data['release_date'].dt.year >= 1980]
```

```python
data.isnull().sum()
```

```python
# CÃ¡c cá»™t chuá»—i sáº½ Ä‘Æ°á»£c thay tháº¿ báº±ng 'None' hoáº·c chuá»—i rá»—ng
data['title'] = data['title'].fillna('Unknown')
data['status'] = data['status'].fillna('None')
data['release_date'] = data['release_date'].fillna('Unknown')
data['adult'] = data['adult'].fillna(False)
data['homepage'] = data['homepage'].fillna('None')
data['imdb_id'] = data['imdb_id'].fillna('None')
data['original_language'] = data['original_language'].fillna('None')
data['tagline'] = data['tagline'].fillna('None')
data['genres'] = data['genres'].fillna('None')
data['production_companies'] = data['production_companies'].fillna('None')
data['production_countries'] = data['production_countries'].fillna('None')
data['spoken_languages'] = data['spoken_languages'].fillna('None')

# CÃ¡c cá»™t sá»‘ thay tháº¿ NaN báº±ng 0 hoáº·c giÃ¡ trá»‹ há»£p lÃ½
data['vote_average'] = data['vote_average'].fillna(0)
data['vote_count'] = data['vote_count'].fillna(0)
data['revenue'] = data['revenue'].fillna(0)
data['runtime'] = data['runtime'].fillna(0)
data['budget'] = data['budget'].fillna(0)
data['popularity'] = data['popularity'].fillna(0)
```

```python
# Náº¿u cá»™t 'budget' cÃ³ giÃ¡ trá»‹ 0, thay tháº¿ báº±ng giÃ¡ trá»‹ trung vá»‹ cá»§a budget
median_budget = data.loc[data['budget'] > 0, 'budget'].median()
data['budget'] = data['budget'].replace(0, median_budget)

# Lá»c doanh thu (revenue) há»£p lÃ½
# data = data[((data['revenue'] >= 0) & (data['status'] != 'Released')) | ((data['revenue'] != 0) & (data['status'] == 'Released'))]

# TÃ­nh lá»£i nhuáº­n (profit) tá»« revenue vÃ  budget
# data['profit'] = data['revenue'] - data['budget']
```

```python
# Chuyá»ƒn genres, production_companies, production_countries, vÃ  spoken_languages thÃ nh danh sÃ¡ch
data['genres'] = data['genres'].apply(lambda x: [genre.strip() for genre in x.split(',') if genre.strip() != 'None'])
data['production_companies'] = data['production_companies'].apply(lambda x: [comp.strip() for comp in x.split(',') if comp.strip() != 'None'])
data['production_countries'] = data['production_countries'].apply(lambda x: [country.strip() for country in x.split(',') if country.strip() != 'None'])
data['spoken_languages'] = data['spoken_languages'].apply(lambda x: [lang.strip() for lang in x.split(',') if lang.strip() != ''])
```

```python
# Thay tháº¿ cÃ¡c giÃ¡ trá»‹ 0 trong 'runtime' báº±ng giÃ¡ trá»‹ trung vá»‹ cá»§a runtime
median_runtime = data.loc[data['runtime'] > 0, 'runtime'].median()
data['runtime'] = data['runtime'].replace(0, median_runtime)

# Äá»‹nh nghÄ©a cÃ¡c khoáº£ng cho 'runtime'
bins = [0, 60, 120, 180, float('inf')]
labels = ['Under 1h', '1h - 2h', '2h - 3h', 'Over 3h']

# Ãp dá»¥ng pd.cut() Ä‘á»ƒ chia cá»™t 'runtime' thÃ nh cÃ¡c khoáº£ng
data['runtime_range'] = pd.cut(data['runtime'], bins=bins, labels=labels, right=False)
```

```python
# PhÃ¢n loáº¡i Ä‘á»™ phá»• biáº¿n (popularity)
popularity_data = data['popularity']
popularity_bins = [0, 0.6, 0.874, 10, 100, float('inf')]
popularity_labels = ['Low', 'Average', 'Good', 'Very Good', 'Excellent']

# PhÃ¢n chia dá»¯ liá»‡u theo cÃ¡c khoáº£ng
data['popularity_category'] = pd.cut(popularity_data, bins=popularity_bins, labels=popularity_labels, include_lowest=True)
```
